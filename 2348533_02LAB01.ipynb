```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/data - data.csv")
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop("street",axis=1,inplace=True)
df.info()
```

```python
df.drop("city",axis=1,inplace=True)
df.drop("statezip",axis=1,inplace=True)
df.drop("country",axis=1,inplace=True)
df.info()
```

```python
df.drop("date",axis=1,inplace=True)
```

```python
df.describe()
```

```python
correlation_matrix = df.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 10))  # Adjust the figure size if needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix', fontsize=14)
plt.show()
```

```python
col_bins=[0,3000,6000,float('inf')]
col_labels=['small_house','medium_house','big_house']
df["sqft_above"]=pd.cut(df["sqft_above"],bins=col_bins,labels=col_labels,right=False)
df["sqft_above"]
```

```python
col_bins=[1900,1940,1980,float('inf')]
col_labels=['old_house','medium_age_house','new_house']
df["yr_built"]=pd.cut(df["yr_built"],bins=col_bins,labels=col_labels,right=False)
df["yr_built"]
```
